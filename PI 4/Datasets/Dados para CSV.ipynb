```python
import pandas as pd

file_names = ['GLOBAL_DATAFLOW_2000-2005.xlsx','GLOBAL_DATAFLOW_2006-2010.xlsx', 'GLOBAL_DATAFLOW_2011-2015.xlsx','GLOBAL_DATAFLOW_2016-2023.xlsx']
df_list = []
for file_name in file_names:
    df = pd.read_excel(file_name)
    df_list.append(df)
df_concat = pd.concat(df_list)

df_concat.to_csv('dados.csv', index=False)

```
